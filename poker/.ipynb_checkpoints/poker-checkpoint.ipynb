## Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid", color_codes=True)
```

```python
df = pd.read_csv("one_dollar_spin_and_go.csv")
```

```python
mylogin = 'fa538846'
```

___
## Exploratory Data Analysis

```python
# Wins and defeats (%)
wd = df.pivot_table(index='tourn_id',columns='name',values='balance',aggfunc='sum')[mylogin]
wd = wd.value_counts().rename(index={-500:'defeat',1000:'win'})
print(pd.DataFrame(wd/wd.sum())[mylogin].round(2).to_string())
```

```python
plt.title('\nChips performance per hands\n',fontsize=16,loc='left')
(
    df
    .query("name == @mylogin")['balance'].reset_index(drop=True)
    .cumsum()
    .plot(kind='line',figsize=(10,5),lw=1)
)
plt.xlabel('hands')
plt.ylabel('chips balance')
plt.show()
```

```python
plt.title('\nChips performance per date\n',fontsize=16,loc='left')
(
    df
    .pivot_table(index='date',columns='name',values='balance',aggfunc='sum')[mylogin]
    .plot(kind='bar',figsize=(16,4))
)
plt.show()
```

```python
plt.title('\nChips performance per blinds levels\n',fontsize=16,loc='left')
(
    df
    .pivot_table(index='level',columns='name',values='balance',aggfunc='sum')[mylogin]
    .plot(kind='bar',figsize=(10,5),color='darkred',rot=0)
)
plt.show()
```

```python
def clean(x):
    return int(x[0:2].replace(':',''))
df['hour'] = df['time'].apply(clean)
df['day'] = pd.to_datetime(df['date']).dt.dayofweek
```

```python
plt.title('\nChips performance per weekday\n',fontsize=16,loc='left')
(
    df
    .pivot_table(index='day',columns='name',values='balance',aggfunc='sum')[mylogin]
    .plot(kind='bar',figsize=(10,5),color='black',)
    .set_xticklabels(['Mon','Tue','Wed','Thu','Fri','Sat','Sun'],rotation=0)
)
plt.xlabel('')
plt.show()
```

```python
plt.title('\nChips performance per hour of the day\n',fontsize=16,loc='left')
(
    df
    .pivot_table(index='hour',columns='name',values='balance',aggfunc='sum')[mylogin]
    .plot(kind='bar',figsize=(16,4),color='darkblue',rot=0)
)
plt.show()
```

```python
plt.subplots(figsize=(5,9))
sns.heatmap(
    df
    .query("name == @mylogin")
    .pivot_table(index='hour',columns='day',values='balance',aggfunc='sum')
    .rename({0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}, axis='columns'),
    cmap='RdYlGn'
)
plt.title('\nChips performance\nper weekday and hour of the day\n',fontsize=16,loc='left')
plt.show()
```

```python
print('\nChips performance\nper cards combinations & blind level')
def cardslevel(x):
    for asc,bw in zip([False,True],['BEST','WORST']):
        dfcards = (
            df
            .query("level == @x")
            .pivot_table(index='combination',columns='name',values='balance',aggfunc='sum')[mylogin]
            .sort_values(ascending=asc)[0:5]
            .dropna()
        )
        print('-'*36,f'\n {bw} FIVE HANDS\n','-'*36,sep='')
        for i,j in zip(dfcards.index,dfcards.values):
            print(i,f'%{35-len(i)}s' % str(int(j)))
    if asc: print('-'*36)
for i in range(1,7):
    print(f'\n\n LEVEL {i}')
    cardslevel(i)
```
